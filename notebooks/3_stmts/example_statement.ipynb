```python
# 좋아하는 영화가 넷플릭스에 존재하는 지 확인하는 반복문 예제

netflix_top10_names = ['유체이탈자','킬러의보디가드2','자산어보','캐시트럭','악인전','사이버지옥','야차','성난황소','기묘한이야기','우리이혼했어요']
prefix = '기묘'

while len(netflix_top10_names) > 0:
    pick = netflix_top10_names.pop()
    if pick.startswith(prefix):
        print("'{}'(으)로 시작하는 영화 '{}'를 찾았습니다".format(prefix, pick))
        break
else:
    print("'{}'로 시작하는 영화가 없습니다".format(prefix))
```

```python
netflix_top10_movies = {
    '유체이탈자': { 
        'age': 15 , 
        'jenre': ['미스터리']
    }, 
    '킬러의보디가드2': { 
        'age': 15 , 
        'jenre': ['액션','코미디']
    }, 
    '자산어보': { 
        'age': 12 , 
        'jenre': ['사극']
    }, 
    '캐시트럭': { 
        'age': 18 , 
        'jenre': ['액션','스릴러']
    }, 
    '악인전': { 
        'age': 18 , 
        'jenre': ['액션','스릴러']
    }, 
    '사이버지옥': { 
        'age': 18 , 
        'jenre': ['범죄','실화']
    }, 
    '야차': { 
        'age': 15 , 
        'jenre': ['액션','스릴러']
    }, 
    '성난황소': { 
        'age': 15 , 
        'jenre': ['액션','스릴러']
    }, 
    '기묘한이야기': { 
        'age': 15 , 
        'jenre': ['SF']
    }, 
    '우리이혼했어요': { 
        'age': 18 , 
        'jenre': ['리얼리티']
    } 
}

# 반복문 조건
#* 입력된 영화제목과 완전일치된 영화가 나오면 종료 (break), 일부 일치된 영화가 있다면 출력하고 계속 (continue)
#* 입력된 나이에 맞지 않는 영화는 스킵 (continue)
#* 정확히 매치되지 않는 영화가 없다면 오류 메시지를 출력 (else)

keyword = '이'

for movie in netflix_top10_movies:
    if movie == keyword:
        print("요청하신 키워드{}(와)과 완전히 일치하는 영화를 찾았습니다 - {}".format(keyword, movie))
        break
    if keyword in movie:
        replaced = movie.replace(keyword, '['+keyword+']')
        print("요청하신 키워드{}(와)과 일부 일치하는 영화를 찾았습니다 - {}".format(keyword, replaced))
        continue
else:
    print("요청하신 키워드{}(와)과 완전히 일치하는 영화를 찾지 못 했습니다".format(keyword))
```
